```python
import requests
import pandas as pd

url = "https://api.spacexdata.com/v4/launches"
response = requests.get(url)
data = response.json()  # this is like a big dictionary
df = pd.json_normalize(data)  # turns it into a table

df.head()  # shows the first few rows
```

```python
# Keep only the useful columns
df = df[['name', 'date_utc', 'rocket', 'launchpad', 'success', 'payloads']]

# Drop rows where success info is missing
df = df.dropna(subset=['success'])

# Convert date column to datetime format
df['date_utc'] = pd.to_datetime(df['date_utc'])

df.head()
```

```python
# Fetch rocket data
rockets_url = "https://api.spacexdata.com/v4/rockets"
rockets = requests.get(rockets_url).json()
df_rockets = pd.json_normalize(rockets)

# Join with your launch data
rocket_dict = {r['id']: r['name'] for r in rockets}
df['rocket_name'] = df['rocket'].map(rocket_dict)
```

```python
# Fetch launchpad data
launchpads_url = "https://api.spacexdata.com/v4/launchpads"
launchpads = requests.get(launchpads_url).json()
df_launchpads = pd.json_normalize(launchpads)

# Create a mapping of launchpad id to name
launchpad_name_dict = {lp['id']: lp['name'] for lp in launchpads}
launchpad_region_dict = {lp['id']: lp['region'] for lp in launchpads}

# Map to your dataframe
df['launch_site'] = df['launchpad'].map(launchpad_name_dict)
df['launch_region'] = df['launchpad'].map(launchpad_region_dict)
```

```python
# Fetch all payloads
payloads_url = "https://api.spacexdata.com/v4/payloads"
payloads = requests.get(payloads_url).json()
df_payloads = pd.json_normalize(payloads)

# Create a dictionary to map payload ID to mass
payload_mass_dict = {p['id']: p['mass_kg'] for p in payloads}

# Extract the first payload ID from the list
df['payload_id'] = df['payloads'].apply(lambda x: x[0] if x else None)
df['payload_mass_kg'] = df['payload_id'].map(payload_mass_dict)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier

# Separate numerical and categorical features
num_features = ['payload_mass_kg', 'year']
cat_features = ['rocket_name', 'launch_site']

# Preprocessing
num_transformer = StandardScaler()
cat_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(transformers=[
    ('num', num_transformer, num_features),
    ('cat', cat_transformer, cat_features)
])

# Define model pipeline
model = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
print(X_train.columns)
print(y_train.head())
```

```python
model.fit(X_train, y_train)
```

```python
import requests
import time

API_KEY = "YOUR_API_KEY_HERE"  # ðŸ”‘ Replace this with your actual API key

def fetch_weather(lat, lon, date):
    unix_timestamp = int(date.timestamp())
    url = f"https://api.openweathermap.org/data/2.5/onecall/timemachine"
    
    params = {
        "lat": lat,
        "lon": lon,
        "dt": unix_timestamp,
        "appid": API_KEY,
        "units": "metric"
    }
    
    try:
        response = requests.get(url, params=params)
        data = response.json()
        current = data.get("current", {})
        return {
            "temp": current.get("temp"),
            "humidity": current.get("humidity"),
            "wind_speed": current.get("wind_speed"),
            "weather_main": current.get("weather", [{}])[0].get("main")
        }
    except Exception as e:
        print(f"Failed for {lat},{lon} @ {date}: {e}")
        return {"temp": None, "humidity": None, "wind_speed": None, "weather_main": None}
```

```python
weather_data = df.head(5).apply(
    lambda row: fetch_weather(row['latitude'], row['longitude'], row['date_utc']), axis=1
)

# Convert results into separate columns
weather_df = pd.DataFrame(weather_data.tolist())
df_sample = pd.concat([df.head(5).reset_index(drop=True), weather_df], axis=1)

df_sample.head()
```
